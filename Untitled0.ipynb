```python
import  tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
fashion_mnist = tf.keras.datasets.fashion_mnist
(train_image,train_label),(test_image,test_label)=fashion_mnist.load_data()
```

```python
print(train_image.shape) # for checking if data is loaded properly 
```

```python
# as we need to print names of object identified, making an array for the same , because output will be in numbers

object_name = ['Top' ,'Trouser' ,'Pullover' ,'Dress' ,'Coat' ,'Sandal' ,'Shirt' ,'Sneaker' ,'Bag' ,'Ankleboot' ]
```

```python
#Lets explore data a bit and understand all its basic information

print(train_image.shape)
print(test_image.shape)
print(len(train_label))
print(len(test_label))

#so after running we can see that train_images have 60000 28x28px images and test_images have 10000 of same size
```

```python
# Preprocess the data

plt.figure()
plt.imshow(train_image[1])
plt.colorbar()
plt.show()

# To inspect pixel range add colorbar

# Pixel range is from 0 to 255 lets change it from 0 to 1 
```

```python
# Pixel range is from 0 to 255 lets change it from 0 to 1 

train_image = train_image/255.0
test_image = test_image/255.0
```

```python
# checking the image if it is in correct shape and is normalised
plt.figure()
plt.imshow(train_image[0],cmap=plt.cm.binary)
plt.colorbar()
plt.show()
```

```python
#Build a model 
model = tf.keras.Sequential([
                             tf.keras.layers.Flatten(input_shape=(28,28)),
                             tf.keras.layers.Dense(128,activation='relu'),
                             tf.keras.layers.Dense(10)
])
#Here flatten is to reformat the images from 28X28 to 784 (2d to 1d)
#First Dense layer has128 nodes with relu activation func , second returns array of 10 logits


#Compile the model

model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
#Train the model so , feed trainind data first here it is train_image and train_labels, model will learn to associate image and labels, then u can ask model to predict the test images
#model.fit fits entire training data into model

model.fit(train_image,train_label,epochs=12)
```

```python
#Evaluate accuracy and make changes accordingly
# as 8 epochs gives <80 % accuracy lets change it to 15 


test_loss,test_accuracy=model.evaluate(test_image,test_label,verbose=2)
print(test_accuracy)
```

```python
#lets take another value of epoch because ig the model overfitted with data 
probability_model=tf.keras.Sequential([model,tf.keras.layers.Softmax()])
predictions = probability_model.predict(test_image)
```

```python
#Lets see the prediction 
print(test_label[0])
print(np.argmax(predictions[0]))
```

```python
# The result is correct last class have the highest probability
```

```python
#Lets make a graph of 10 to 15 test_images with their predictions 

def plot_image(i,predictions,test_label,img):
  test_label,img=test_label[i],img[i]
  plt.imshow(img,cmap=plt.cm.binary)
  predicted_label=np.argmax(predictions)
  #to highlight right and wrong predicted labels
  if predicted_label == test_label:
    color='blue'
  else:
    color='red'
  plt.xlabel("{} {:2.0f}%({})".format(object_name[predicted_label],
                                      100*np.max(predictions),
                                      object_name[test_label]),
                                      color=color)

def plot_value(i,predictions,test_labels):
  test_label=test_labels[i]
  plt.xticks(range(10))
  plt.yticks([])
  plot=plt.bar(range(10),predictions,color="#777777")
  plt.ylim=([0,1])
  predicted_label=np.argmax(predictions)
  plot[predicted_label].set_color('red')
  plot[test_label].set_color('blue')
```

```python
i=0
plt.figure(figsize=(6,3))
plt.subplot(1,2,1)
plot_image(i,predictions[i],test_label,test_image)
plt.subplot(1,2,2)
plot_value(i,predictions[i],test_label)
plt.show()
```

```python
#plotting several images 

num_rows=4
num_columns=4
num_images=num_rows*num_columns
plt.figure(figsize=(2*2*num_columns,2*num_rows))
for i in range(num_images):
  plt.subplot(num_rows,2*num_columns,2*i+1)
  plot_image(i,predictions[i],test_label,test_image)
  plt.subplot(num_rows,2*num_columns,2*i+2)
  plot_value(i,predictions[i],test_label)
plt.tight_layout()
plt.show()
```
